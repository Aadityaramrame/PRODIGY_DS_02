```python
import pandas as pd
import numpy as np
```

```python
url = "https://github.com/m4gaikwad/PandasTutorial/raw/main/online_store_customer_data.csv"
data = pd.read_csv(url,on_bad_lines='skip')
data
```

```python
data.rename(columns={"Transactiond_Date" : 'Date',
                     'State_name ': 'State'},inplace ='True')
```

```python
data.sample(5)
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(keep = 'first',inplace=True)
```

```python
data.iloc[1:5,2:6]
```

```python
data[data['State_names']== 'Kansas' ]
```

```python
data[(data['Age'] >= 45) & (data['Age']<= 60)]
```

```python
data[(data['Age'] >= 45) & (data['Age']<= 60) & (data['Marital_status'] == "Single" )& (data["Gender"] == 'Male')]
```

```python
data['Age'].isna().sum()
```

```python
df = data.dropna(subset = ['Age'])
```

```python
df['Age'].isna().sum()
```

```python
df['Age_Range'] = np.where(df['Age'] <= 30,'Young',
                           np.where(df['Age']<= 45,'Middle',
                                    'old'))
df['Age_Range']
```

```python
df.head(5)
```

```python
df.groupby('Age_Range')['Age'].count()
```
